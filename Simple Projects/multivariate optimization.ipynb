## Mathematical Optimization
- Continuation of previous optimization notebook, using multivariable functions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Function definition

$$f(x, y) = e^{-\frac{x}{2}^2} \cdot e^{-\frac{y}{2}^2} $$


## Optimization

```python
from scipy import optimize

def gaussian_mixture(x):
    return np.exp(-(x[0]/2)**2) * np.exp(-(x[1]/2)**2)

x, y = np.meshgrid(
    np.linspace(-5, 5, 1000), 
    np.linspace(-5, 5, 1000)
)


z = gaussian_mixture([x, y])
result = optimize.minimize(gaussian_mixture, x0=np.zeros((2,)))
print(result)
```

# Plots

```python
plt.figure(figsize=(8, 4))
ax = plt.subplot(1, 2, 1)
ax.imshow(z, cmap="inferno", extent=[-5, 5, 5, -5])
ax.scatter(result.x[0], result.x[1], color='lime')
plt.title("Heat Map")

ax = plt.subplot(1, 2, 2, projection='3d')
ax.plot_surface(x, y, z, cmap="inferno",
                       linewidth=0, antialiased=False)
ax.scatter(result.x[0], result.x[1], -1,  color='lime')
plt.title("3-Dimensional Map")
```

```python

ax = plt.subplot(1, 2, 1)
contour_levels = np.linspace(0, np.max(z), 10)
ax.contour(x, y, z, levels=contour_levels, cmap="inferno", alpha=0.5)
ax.scatter(result.x[0], result.x[1], color='lime')
plt.title("Contour Map");
```
