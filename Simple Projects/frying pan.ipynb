# Frying Pan with Heat

- Using convolution and numerical programming

```python
import numpy as np
from scipy.signal import convolve2d

import matplotlib.pyplot as plt
import matplotlib.animation as anim
%matplotlib notebook
```

## Initializing stove and cold pan

### Heat equation 

$T = \alpha(f(x, y))$

```python
environment = np.zeros((100,100)) # cold pan
environment[-1, :] = 1.0 # heat from the stove
```

```python
diffusion_kernel = np.array([[0, 1, 0],
                 [1, -4, 1],
                 [0, 1, 0]])
alpha = 0.1

def convolution(n_frames):
    dT = convolve2d(environment, diffusion_kernel, mode='same')
    environment[...] += dT * alpha
    environment[-1, :] = 1.0
    image.set_data(environment)
```

```python
fig, ax = plt.subplots()
image = ax.imshow(environment, cmap='inferno')

def update_plot(frame):
    convolution(environment)
    image.set_data(environment)
    
animation = anim.FuncAnimation(fig, update_plot, interval=30)
plt.show()
```
