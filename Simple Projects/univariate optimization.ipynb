# Mathematical Optimization
- Univariate function optimization, plots, and constraints
- Practice of optimizaiton using SciPy and NumPy libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Function Definition

$f(x)={cos(x) \cdot e^{-\frac{3(x - 0.134)^2}{10}}}$

```python
def f(x):
    return np.cos(x) * np.exp(-0.3*(x-0.134)**2)
```

## Plotting

```python
x = np.linspace(-10,10,100)
plt.plot(x,f(x), 'black')
plt.grid()
```

## Function Evaluation

```python
x_rand = np.random.uniform(-10,10,20)
plt.plot(x_rand, f(x_rand), '.r')
plt.grid()
```

## Scalar Optimization

```python
from scipy import optimize

result = optimize.minimize_scalar(f)
print(result)

plt.plot(x, f(x), "black")
plt.plot(result.x, f(result.x), 'ro')
plt.grid()
```

## Local Minima

```python
result = optimize.minimize(f, x0=-5.0) # local minima closest to x=-5
print(result)

plt.plot(x, f(x), "black")
plt.plot(result.x, f(result.x), 'ro')
plt.grid()
```

## Inequality Constraint

$f(x)>0$

```python
constraints = [
 {'type':'ineq','fun': lambda x: x},   
]

result = optimize.minimize(f, x0=2.5,  constraints=constraints)
print(result)

plt.plot(x, f(x), 'black')
plt.plot(result.x, f(result.x), 'ro')
plt.grid()
```

  ## Equality Constraint

$f(x)=0$

```python
def constraint_f(x):
    return np.sin(x)

constraints = [
    {'type':'eq','fun': constraint_f},   
]

result = optimize.minimize(f, x0=-1,  constraints=constraints)
print(result)

plt.plot(x, f(x), 'black')
plt.plot(result.x, f(result.x), 'ro')
plt.plot(x, constraint_f(x), 'b')
plt.grid()
```
