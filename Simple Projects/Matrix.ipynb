# Matrices 
- Practice linear algebraic functions and understanding using NumPy, and MatPlotLib

## Initializing Matrices and Matrix Multiplication

```python

A = np.random.rand(4, 2)
B = np.random.rand(2, 4)
C = np.dot(A, B)
print(C)
```

### Plotting Output Matrix

```python
# Plotting heatmap
plt.imshow(C, cmap='inferno')
plt.colorbar()
plt.show()

# Plotting 3D surface map
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x, y = np.meshgrid(range(4), range(4))
ax.plot_surface(x, y, C, cmap='inferno')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

# Matrix scatter plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(range(16), C.flatten(), c=C.flatten(), cmap='inferno')
ax.set_xlabel('Index')
ax.set_ylabel('Value')
plt.show()

# Matrix 3D Contour Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(4)
X, Y = np.meshgrid(x, y)
Z = C
ax.contourf(X, Y, Z, cmap='plasma')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## Singular Value Decomposition of Output Matrix

```python
U, S, V = np.linalg.svd(C)

print("U:", U)
print("S:", S)
print("V:", V)

# Plotting singular values for U, S, and V
plt.plot(U)
plt.xlabel('Singular Value (U)')
plt.ylabel('Magnitude')
plt.show()

plt.plot(S)
plt.xlabel('Singular Value (S)')
plt.ylabel('Magnitude')
plt.show()

plt.plot(V)
plt.xlabel('Singular Value (V)')
plt.ylabel('Magnitude')
plt.show()
```

## Eigenvalue Decomposition of Output Matrix

```python
eigenvalues, eigenvectors = np.linalg.eig(C)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:", eigenvectors)

plt.plot(eigenvalues)
plt.xlabel('Eigenvalue')
plt.ylabel('Magnitude')
plt.show()
```

## Matrix Inverse

```python
output_inverse = np.linalg.inv(C)
print(output_inverse)

plt.imshow(output_inverse, cmap='inferno')
plt.colorbar()
plt.show()
```
