***This is a simple linear regression model run on voting data, specifically on how voting percentage of a given candidate is predicted by their voting percentage***




**The data set that is used contains the following variables**

    1. state                    state postal code
    2. district                 congressional district
    3. democA                   =1 if A is democrat
    4. voteA                    percent vote for A
    5. expendA                  campaign expends. by A, $1000s
6. expendB                  campaign expends. by B, $1000s
    7. prtystrA                 % vote for president
    8. lexpendA                 log(expendA)
    9. lexpendB                 log(expendB)
    10. shareA                  100*(expendA/(expendA+expendB)) 
   
  


```python
# Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading in the dataset using pandas, and portraying the head
df = pd.read_excel("C:/Users/tyler/OneDrive/Desktop/Classes/Econometric Methods/Data/Data Sets-Excel/vote1_updated.xls")
print(df.head())
```

```python
# Identifying the independent and dependent variables for the linear regression
# Predicting the voting percentage based off the expenditure for the candidate A
X = df[['expendA']]
y = df[['voteA']]
```

```python
# Splitting the data into train and test sets, using the Sci-Kit Learn library
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
X_train = X_train.values.reshape(-1,1) # Reshaping the array for the linear regressor
```

```python
# Fitting the linear regression to the training set
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Predicting the results of the test
y_pred = lr.predict(X_test)
```

```python
# Visualizations for training set
plt.scatter(X_train, y_train, color = "red") # Plotting the obersvations
plt.plot(X_train, lr.predict(X_train), color = "blue") # Plotting the regression
plt.title("Voting Percentage vs. Expenditure (Training Set)")
plt.xlabel("Campaign expenditure by $1000")
plt.ylabel("Voting percentage of candidate")
plt.show()
```

```python
# Visualizations for the test set
plt.scatter(X_test, y_test, color = "red") # Plotting the obersvations
plt.plot(X_train, lr.predict(X_train), color = "blue") # Plotting the regression
plt.title("Voting Percentage vs. Expenditure (Test Set)")
plt.xlabel("Campaign expenditure by $1000")
plt.ylabel("Voting percentage of candidate")
plt.show()
```

**As expected, there seems to be a seemingly positive linear relationship between voting expenditure and voting percentage for a given candidate, however looking at both the training and test sets, and the data in general there tends to be quite a few outliers. Nevertheless, using the various libraries of Python, and excel data on voting, meaningful relationships in data can be analyzed and viewed.**
